```python
#working with stackplot
days = [1,2,3,4,5]
sleeping = [6,7,9,3,2]
eating =   [3,4,2,5,7]
working =  [8,7,6,5,3]
playing =  [5,9,3,5,6]
#to add labels we must plot them
plt.plot([],[],label='sleeping',color='m', linewidth=6)
plt.plot([],[],label='eating',color='r',linewidth=6)
plt.plot([],[],label='working',color='b',linewidth=6)
plt.plot([],[],label='playing',color='k',linewidth=6)


plt.stackplot(days, sleeping,eating,working,playing,colors=['m','r','b','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('graphic')
plt.show()
```
